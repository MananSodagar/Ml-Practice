```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
```

```python
df=pd.read_csv("C:/Users/manan/Desktop/ML dataset/HR_comma_sep.csv")
```

```python
df.head()
```

```python
dummysalary=pd.get_dummies(df.salary,prefix="salary")
```

```python
df2=pd.concat([df,dummysalary],axis=1)
df2.head()
```

```python
df2.drop(['salary_low','salary'],axis=1,inplace=True)
```

```python
df2.head()
```

```python
X=df2[['satisfaction_level','average_montly_hours','promotion_last_5years']]
```

```python
X
```

```python
y=df2.left
```

```python
X_t,X_v,y_t,y_v=train_test_split(X,y,random_state=43,test_size=.33)
```

```python
model=LogisticRegression(C=0.1)
model.fit(X_t,y_t)
```

```python
model.predict(X_v)
```

```python
model.score(X_v,y_v)
```

```python
model2=SVC()
```

```python
model2.fit(X_t,y_t)
model2.predict(X_v)
model2.score(X_v,y_v)
```
