```python
df2=df[(df.price>min_price) & (df.price<max_price)]
df2.price.describe()
```

```python
df2.price.hist(bins=1000)
```

```python
upper_limit = df2.price.std()*3
lower_limit=df2.price.mean()*-3*df2.price.std()
```

```python
df2[(df2.price<lower_limit) | (df2.price>upper_limit)]
```

```python
df2.price.hist(bins=100)
```

```python
df3=df2[(df2.price>lower_limit) & (df2.price<upper_limit)]
```

# Using Z-score value

```python
df2["Zscore1"]=(df2.price-df2.price.mean())/df2.price.std()
```

```python
df2.Zscore1.describe()
```

```python
df4=df2[(df2.Zscore1>-2) & (df2.Zscore1<2)]
```

```python
df4.price.hist(bins=100)
```
